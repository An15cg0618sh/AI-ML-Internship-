```python
df = pd.read_csv("Titanic-Dataset.csv", na_values=[" ", ""], keep_default_na=False)
print(np.unique(df["Age"]))
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
print(np.unique(df["Age"]))
```

```python
df['Age'] = df['Age'].round().astype('Int64') # Use Int64 to handle NaN
print(df.dtypes['Age'])
```

```python
df['Embarked'] = df['Embarked'].replace('nan', np.nan)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'] = df['Embarked'].astype('category')
print(df['Embarked'].unique())
```

```python

df = pd.read_csv("Titanic-Dataset.csv")
df.drop('Cabin', axis=1, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = 1  # Initialize to yes
df['IsAlone'].loc[df['FamilySize'] > 1]
```

```python
numerical_features = df.select_dtypes(include=np.number).columns
numerical_df = df[numerical_features]

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
